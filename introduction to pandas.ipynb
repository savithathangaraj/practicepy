```python
import pandas as pd
car=pd.read_csv("car-sales.csv")
seats_column=pd.Series([5,5,5,5,5])
car["seats"]=seats_column
car
```

```python
car["seats"]=car["seats"].fillna(value=5,inplace=False)
car
```

```python
fuel_economy=pd.Series([7.5,9.2,5.0,6.7,8.3,7.3,4.5,9.2])
car["economy"]=fuel_economy
car["total fuel used(L)"]=car["Odometer (KM)"]/100*car["economy"]
car["no of wheels"]=4
car
```
